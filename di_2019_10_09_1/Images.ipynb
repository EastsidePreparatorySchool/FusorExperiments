## Appendix: Ways to embed local images

### First from a markdown cell:

directly from markdown
![local image](IMG_6899.jpg)

Note how you cannot rotate pictures that your phone took in a different orientation. Annoying. 

### Now from embedded HTML in Markdown:

from HTML in markdown
<img style="transform:rotate(270deg);width:800px" src="img_6899.jpg" />

you can see that the rotate works correctly. It is important to put the picture into its own cell, 
otherwise the rotation will make it go across the boundaries of adjoining text.
